### Analysis of temporal patterns in animal movement networks

```python
import pandas as pd
import traja
```

##### Dataset 1 contains the movement of bumblebee tracked with a harmonic radar : Bumblebee search trajectory

```python
dataset1 = pd.read_csv('pasqueretta_/dataset_S1.csv')
dataset1.head()
```

```python
dataset1.ID.unique()
```

```python
dataset1.plot.line(x = 'Longitude',y = 'Latitude',figsize=(10,5), grid=True, label = 'Trajectory')
```
